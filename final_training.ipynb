```python
from ultralytics import YOLO

# Load your YOLOv8 model
model = YOLO('yolov8n.pt')

# Define hyperparameters
learning_rate = 0.001
batch_size = 32
image_size = 640
epochs = 100

# Train the model with updated parameters
model.train(
    data='/content/PPE-2/data.yaml',
    epochs=epochs,
    imgsz=image_size,
    batch=batch_size,
    lr0=learning_rate,
    lrf=0.01,
    augment=False,
    project='runs/train',
    name='experiment_lr{}_bs{}_epochs{}_imgsz{}'.format(learning_rate, batch_size, epochs, image_size),
    patience=10
)
```

```python
model.val(data='/content/PPE-2/data.yaml')
```
